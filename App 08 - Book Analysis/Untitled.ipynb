```python
import re
```

# Load the Book

```python
with open("miracle_in_the_andes.txt", "r", encoding="utf-8") as file:
    book = file.read()
```

# Extract the paragraphs where "love" was used

```python
pattern = re.compile("[\n]{1}[^\n]*[^a-zA-Z]+love[^a-zA-Z][^\n]*\n")
findings = re.findall(pattern, book)
findings
```

# Extract Chapter Titles

```python
pattern = re.compile("Chapter [0-9]+[\n]{2}(.*)")
findings = re.findall(pattern, book)
findings
```

# Function that finds the occurance of any word

```python
def find(word):
    pattern = re.compile(f"{word}[ .,]")
    findings = re.findall(pattern, book.lower())
    return len(findings)
```

# Call the Function

```python
find("hate")
```

```python
import pandas as pd
movies = pd.read_csv("movies.csv", sep=",")
movies.head()
```

# Content-Based Filtering

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words="english")

movies["overview"] = movies["overview"].fillna("")
```

```python
tfidf_matrix = tfidf.fit_transform(movies["overview"])
```

```python
pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names())
```

```python
tfidf_matrix.shape
```

### Similarity Matrix

```python
from sklearn.metrics.pairwise import linear_kernel
similarity_matrix = linear_kernel(tfidf_matrix,tfidf_matrix)
similarity_matrix
# Compares each of the movies with each other, for a value of 1 it is being compared with itself so it 
# gets the highest value. 
```

### Find similar movies compared to a specific movie

```python
movie_title = "John Carter"
```

```python
idx = movies.loc[movies["title"]==movie_title].index[0]
idx
```

```python
scores = list(enumerate(similarity_matrix[idx]))
scores
```

```python
scores = sorted(scores, key=lambda x: x[1], reverse=True)
scores
```

```python
movie_indices = [tpl[0] for tpl in scores[1:4]]
movie_indices
```

```python
list(movies["title"].iloc[movie_indices])
```

```python
def similar_movies(movie_titles, nr_movies):
    idx = movies.loc[movies["title"]==movie_titles].index[0]
    scores = list(enumerate(similarity_matrix[idx]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)
    movie_indices = [tpl[0] for tpl in scores[1:nr_movies+1]]
    similar_titles = list(movies["title"].iloc[movie_indices])
    return similar_titles
```

```python
similar_movies("Kung Fu Panda 3", 5)
```
